# Basic operations

Sometimes data don't come in a convenient format. Let's see how to make some useful transformations.

## Create columns

As an example, we will turn Geolocation into two columns - lat and lon
A few steps here... first, we remove the non-numeric characters '(' and ')'. Then we split the string into two columns. Then we convert to numeric and check our work.

```
# look at the first 10 rows and scroll over - it's a string ('object')
print(df["Geolocation"].dtype)
print(df["Geolocation"][0])
```

I will create an artificial column first, just in case I do some mistake and have to run the whole thing again.

```
df['Geolocation2'] = df['Geolocation']
df.head(n=5)
```

Next, let's get rid of "string stuff", we want numbers here...

```
df['Geolocation2'] = df['Geolocation2'].str.replace('(', '')
df['Geolocation2'] = df['Geolocation2'].str.replace(')', '')
df['Geolocation2'] = df['Geolocation2'].str.replace(',', '')
df.head() # check your work
```

We can see how things have changed...

```
print("Before:",df["Geolocation"][0], "After:",df["Geolocation2"][0])
```

## Split column

Now we will take the content of one column (pair of numbers) and split into two.

```
# let's make a new column for latitude and longitude (the geographic coords)
# Use expand=True to create two columns; otherwise, you will get a dataframe with a single column, with each entry being a list
new = df['Geolocation2'].str.split(" ", expand = True)
new
```

And that's how you create columns. Easy, right?

```
df['lat'] = new[0] # first element of the string
df['lon'] = new[1] # second element of the string
df.head()
```

Check your work!!! You will see that we are not quite there yet...

```
# check column types
df.info()

# uh oh! still stored as a string!
```

## Change the type of a column

This feature is **super** important. Lat-lon coordinates are not so useful as strings, but things change if they are converted to numbers, so let's do that.

For more details, see https://pandas.pydata.org/docs/reference/api/pandas.to_numeric.html

```
# lat to numeric (coerce will set invalid values as NaN)
df['lat'] = pd.to_numeric(df['lat'], errors='coerce')
# lon to numeric
df['lon'] = pd.to_numeric(df['lon'], errors='coerce')

df.info()
df.head()
```

## Delete column (a different approach)
